```python
import pandas as pd
df = pd.read_csv('activities.csv')
df['start_day'] = pd.to_datetime(df['start_time']).dt.date
df.head()
```

```python
import datetime
df.loc[df['start_day'] == datetime.date(2020,2,1)]
```

```python

import datetime

distance_filter = (df['total_distance'] < 2100) & (df['total_distance'] > 1900)
tempo_filter = df['total_timer_time'] < 500
date_filter = (df['start_day'] < datetime.date(2015,11,1)) & (df['start_day'] > datetime.date(2015,3,1))
sorted_by_date = df.sort_values('start_time', ascending=True)
intervals = sorted_by_date[distance_filter & date_filter]
intervals_2 = intervals[intervals['lap_number'] == 2]
intervals_3 = intervals[intervals['lap_number'] == 3]
intervals_4 = intervals[intervals['lap_number'] == 4]
intervals_5 = intervals[intervals['lap_number'] == 5]
```

```python
intervals.head(100)
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
formatter = mdates.DateFormatter("%Y-%m-%d")
ax = plt.gca()
ax.xaxis_date()
ax.xaxis.set_major_formatter(formatter)
plt.rcParams["figure.figsize"] = (24,9)
plt.plot(intervals_2['start_day'], intervals_2['total_timer_time'], label="2")
plt.plot(intervals_3['start_day'], intervals_3['total_timer_time'], label="3")
plt.plot(intervals_4['start_day'], intervals_4['total_timer_time'], label="4")
plt.plot(intervals_5['start_day'], intervals_5['total_timer_time'], label="5")
plt.xticks(intervals['start_day'][::50], rotation=90)
plt.legend()
plt.show()
```

```python

avg_by_day = intervals[['start_day', 'total_timer_time']].groupby(by='start_day', as_index=False).mean()
```

```python
#avg_by_day
avg_by_day['rolling_avg'] = avg_by_day['total_timer_time'].rolling(3, win_type=None).mean()
plt.plot(avg_by_day['start_day'], avg_by_day['rolling_avg'], label="avg")
```
